## 可视化变量分布

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style="darkgrid")
```

### 1. 直方图

```python
x = np.random.normal(size=10000)
sns.distplot(x)
```

```python
# 删除密度曲线并添加一个rug plot，会在每个观察值上画一个小的垂直刻度
sns.distplot(x, kde=False, rug=True)
```

### 2. 核密度估计（KDE图）
和直方图一样，KDE图沿另一个轴的高度，编码一个轴上的观测密度

```python
sns.distplot(x, hist=False, rug=True)     # 默认kde=True，hist=False时不显示直方图
```

```python
sns.kdeplot(x, shade=True)     # 可以直接用kdeplot函数绘制
```

```python
# 观察bw参数对拟合的影响
sns.kdeplot(x)      # bw默认值使用常用规则猜测好的值, 由下图可知默认值是0.2
sns.kdeplot(x, bw=.2, label='bw = 0.2')
sns.kdeplot(x, bw=2, label='bw = 2')
```
