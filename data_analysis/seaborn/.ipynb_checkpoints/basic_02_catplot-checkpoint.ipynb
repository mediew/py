# Seaborn basic 01
## catplot -- 类型图

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style='ticks', color_codes=False)
```

```python
tips = sns.load_dataset('tips')
# catplot 默认用stripplot模式绘图
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# 可以用命名关键字参数jitter控制点在图上分布的宽度
sns.catplot(x='day', y='total_bill', jitter=False, data=tips)
```

```python
# 也可用参数kind指定用swarm模式绘图
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

```python
# 添加字段划分不同的颜色
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
```

如果分类字段（x轴）是数字，将自动按照数字大小排列

```python
sns.catplot(x='size', y='total_bill', kind='swarm', data=tips.query("size != 3"))
# 注意：size不需要加引号
```

```python
# 添加order参数来指定x轴上各标签的顺序
sns.catplot(x='smoker', y='tip', order=['No', 'Yes'], kind='swarm', data=tips)
```

```python
# 要画水平方向的轴，只需交换x，y的数据即可
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

## Boxplots -- 查看分类数据的分布情况
一个矩形盒，两端边的位置分别对应数据的上下四分位数（Q3和Q1）。
在矩形盒内部中位数（Xm）位置画一条线段为中位线。
在Q3+1.5IQR(IQR为四分位距)和Q1－1.5IQR处画两条与中位线一样的线段，
这两条线段为异常值截断点，称其为内限；在Q3+3IQR和Q1－3IQR处画两条线段，称其为外限。
处于内限以外位置的点表示的数据都是异常值，
其中在内限与外限之间的异常值为温和的异常值（mild outliers），在外限以外的为极端的异常值(extreme outliers)。四分位距IQR=Q3-Q1

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
# 添加hue字段以区分不同颜色
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```

```python
# 如果不想两种不同的颜色挨在一块，可以使用dodge参数设置
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.catplot(x="day", y="total_bill", hue="weekend",
            kind="box", dodge=False, data=tips);
```

```python
# 更好的显示数据分布的图 -- boxenplot
diamonds = sns.load_dataset("diamonds")
sns.catplot(x="color", y="price", kind="boxen",
            data=diamonds.sort_values("color"));
```

violinplots -- 模拟核心区域的数据密度

```python
sns.catplot(x="total_bill", y="day", hue="time",
            kind="violin", data=tips);
```

```python
sns.catplot(x="total_bill", y="day", hue="time",
            kind="violin", bw=.15, cut=0,
            data=tips);
```

```python
# 运用split参数将两个分类字段结合在数轴的两边
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="violin", split=True, data=tips)
```

```python
# 在小提琴内部显示每个观察值
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="violin", inner="stick", split=True,
            palette="pastel", data=tips);
```

```python
# 把swarm，strip，box和violin结合起来
g = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=g.ax);
```

## Barplots -- 估计数据的集中趋势


```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic);
```

## countplot -- 显示数据出现的次数

```python
# 兼具条形图和直方图的特点
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic)
```

```python
sns.catplot(y="deck", hue="class", kind="count",
            palette="pastel", edgecolor=".6",
            data=titanic)
```

## Pointplots -- 点图

绘制元素包括参数的估计和置信区间，相同的hue间用线连接起来

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
```

```python
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic);
# 用palette调色板设置颜色，用marker设置标记点样式，用linestyle设置线条样式
```

```python
iris = sns.load_dataset("iris")
sns.catplot(data=iris, orient="h", kind="box");

```

```python
# 不使用data字段引入数据
sns.violinplot(x=iris.species, y=iris.sepal_length);
```

```python
# 设置尺寸
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c");
```

## catplot建立在Facetplot类之上

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips);
```

```python
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log");     # 已对数尺度显示x轴
```
