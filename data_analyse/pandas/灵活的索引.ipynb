#### 列的聚合函数

```python
df['访客数'].mean()    #计算均值
df['访客数'].std()     #计算标准差
df['访客数'].median()  #计算中位数
df['访客数'].max()     #计算最大值
df['访客数'].min()     #计算最小值

print('访客数均值：',df['访客数'].mean())
print('转化率均值：',df['支付转化率'].mean())
print('客单价均值：',df['客单价'].mean())
```

### 场景四：对于流量渠道的数据，我们真正感兴趣的是优质渠道，有哪些渠道的访客数、转化率、客单价都高于平均值呢？他的销售占比又是怎么样的呢？

#### 判断单列是否大于均值

```python
df['访客数'] > df['访客数'].mean()
```

#### 判断是否同时大于均值

```python
(df['访客数'] > df['访客数'].mean()) & (df['支付转化率'] > df['支付转化率'].mean()) & (df['客单价'] > df['客单价'].mean())
```

#### 基于判断结果，筛选出我们希望的优质渠道

```python
df.loc[(df['访客数'] > df['访客数'].mean()) & 
      (df['支付转化率'] > df['支付转化率'].mean()) & 
      (df['客单价'] > df['客单价'].mean()),:]
```
